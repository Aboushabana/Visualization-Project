### When are users using the service the most during the week

```python
sns.countplot(data = df, x = 'start_time_weekday', hue = 'user_type')
plt.title('Users Weekly Activity')
plt.xlabel('Day')
plt.ylabel('Number of Users');
```

## (Visualization 3)

### When are users  using the service the most during the year.

```python
the_order = df.start_time_month.value_counts().index
sns.countplot(data = df, x = 'start_time_month', hue = 'user_type', order=the_order)
plt.xticks(rotation=90)
plt.title('Users Monthly Activity')
plt.xlabel('Month')
plt.ylabel('Number of Users');
```

## (Visualization 4)

### Which type of customers rides bikes for longer time and how old are they

```python
g = sns.FacetGrid(data = df, hue = 'user_type', hue_order = ['Customer', 'Subscriber'], size=8, aspect=2, 
                  palette='viridis_r')
g.map(sns.regplot, 'duration_hour', 'age', x_jitter=0.5, fit_reg = False);
g.add_legend()
plt.title('Users Behavior')
plt.xlabel('Trip Duration in Hours')
plt.ylabel('Users Age');
```
